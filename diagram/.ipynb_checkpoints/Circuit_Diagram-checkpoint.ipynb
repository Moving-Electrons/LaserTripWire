```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
import SchemDraw as schem
import SchemDraw.elements as e
```

```python
d = schem.Drawing()

left = {'cnt':3,
       'labels': ['D3', 'D5', 'EN']}
right = {'cnt': 3,
        'labels': ['GND', 'A2', 'V33'],
        'loc':[0.2, 0.4, 0.8]}

argon = e.blackbox(d.unit*1.5, d.unit*2.5, leadlen=1, linputs=left, rinputs=right, mainlabel='Particle \nArgon')

ic = d.add(argon)
d.add(e.LINE, d='down', l=d.unit*2, xy=ic.GND)
BOT = d.add(e.GND) #Defining Ground element at the bottom.

d.add(e.LINE, d='left', xy=ic.EN, l=d.unit*2.5)
d.add(e.SWITCH_SPST, flip=True, d='down', toy=BOT.start, label='Toggle \nSwitch')
d.add(e.LINE, d='right', tox=BOT.start)

d.add(e.LINE, d='left', xy=ic.D5, l=d.unit*1.5)
d.add(e.BUTTON, flip=True, d='down', toy=BOT.start, label='Push \nButton')
d.add(e.DOT)

d.add(e.DIODE, d='down', xy=ic.D3, label=['+','Photo \nTransistor', '-'])
d.add(e.RES, d='down', toy=BOT.start, label='220 $\Omega$')
d.add(e.DOT)

d.add(e.LINE, d='right', xy=ic.V33, l=d.unit*1.5)
d.add(e.LED, flip=True, d='down', label=['+', 'LED', '-'])
d.add(e.RES, d='down', toy=BOT.start, label='1 $K\Omega$')
d.add(e.LINE, d='left', tox=BOT.start)
d.add(e.DOT)

d.add(e.LINE, d='right', xy=ic.A2, l=d.unit*1.5)
d.add(e.DOT)


d.draw()

```

```python
d.save('laser_trip_wire-2.svg')
```
